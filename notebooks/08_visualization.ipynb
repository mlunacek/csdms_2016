# Visualization

￼￼There are so many great options for visualization.  Here are a few!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import utils

box_data = pd.read_csv('data/boxplot.csv')
```

## Box plots 

```python
_ = box_data.plot(kind="box")
```

### Seborn 

```python
import seaborn as sns

_ = sns.boxplot(data=box_data, color="steelblue")
```

### Plotly 

We'll use cufflinks, a sweet wrapper for plotly via pandas.

```python
import cufflinks as cl
cl.go_offline()

box_data.set_index('house').iplot(kind='box')
```

### Bokeh 

```python
from bokeh.charts import BoxPlot, output_notebook, show, defaults

output_notebook()
```

Bokeh likes it stacked...

```python
stacked = pd.melt(box_data, value_vars=['nosolar', 'solar'], id_vars=['house']).dropna()
stacked.head()
```

```python
p = BoxPlot(stacked, values='value', label='variable')
_ = show(p)
```

### ggplot 

```python
from ggplot import ggplot, aes, geom_boxplot

ggplot(stacked, aes(x='value', y='variable')) + geom_boxplot()
```
